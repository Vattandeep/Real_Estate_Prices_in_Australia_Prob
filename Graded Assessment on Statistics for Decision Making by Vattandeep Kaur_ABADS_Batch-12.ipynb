3. For the suburb Abbotsford, what is the probability that out of 10 properties sold, 
3 will not have car parking? Use the column car in the dataset. Round off your answer to 3 decimal places. 

```python
# For Abbotsford suburb, collecting the data of Abbotsford only
df_abts = dfnew[dfnew['Suburb'] == 'Abbotsford']
```

```python
df_abts.head()
```

```python
df_abts.info()
```

```python
# For the probability that out of 10 properties sold, 3 will not have car parking
# First we need to find out the probablity of having 0 car parking in whole Abbotsford.
Total_properties = len(df_abts)
Total_properties
```

```python
zero_carparking = len(df_abts[df_abts['Car'] == 0])
zero_carparking
```

```python
# Probability of zero car parking in whole Abbotsford
prob_zero_car_parking = zero_carparking/Total_properties
prob_zero_car_parking
```

```python
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
# Probablity that out of 10 properties sold, 3 without parking
# Here Total properties sold (Total Outcomes), 
n = 10
# Properties without car parking (Favorable Outcomes), 
k = 3
# Probability formula - 
pmf = binom.pmf(k, n, prob_zero_car_parking)
pmf
```

```python
# Rounding off the result to 3 decimal places
Result_rdoff = round(pmf,3)
Result_rdoff
```
